```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#let's create a image
im=np.arange(4)
print(f"array is :{im} and shape is :{im.shape}")
```

matriz array.png

```python
im = im[np.newaxis,:] # numpy.newaxis is used to increase the dimension of the existing array by one more dimension, when used once.
print(im)
print(im.shape)
```

```python
im=np.repeat(im,10,axis=0)  # array repeating itself 10 times
print(im)
print(im.shape)
```

```python
plt.imshow(im, cmap='gray')
plt.show()
```

```python
#create another image
new_img = np.arange(256)
new_img = new_img[np.newaxis, :]
new_img = np.repeat(new_img, 100, axis=0)
plt.imshow(new_img, cmap='gray')
plt.show()
```

```python
#create another image
new_img = np.arange(15)
new_img = new_img[np.newaxis, :]
new_img = np.repeat(new_img, 10, axis=0)
plt.imshow(new_img, cmap='gray', vmin=3, vmax=100) # for white =100 for black =3
plt.show()
```

**GrayScale Image in OpenCv**

```python
import cv2
```

```python
#Read the image
img = cv2.imread(r'albert-einstein_gray.jpg', cv2.IMREAD_GRAYSCALE)
print(f"shape is :{img.shape} and type is :{type(img)} and data type is :{img.dtype} ")
plt.imshow(img, cmap='gray')
plt.show()
```

```python
#we have seen that the image through numpy is read only but with cv we can modifify
img[330:430,300:400] = 0 #for left eye
img[330:430,440:540] = 0 #for right eye
#Display the image through matplotlib
plt.imshow(img, cmap='gray')
```

```python
#Display the image through opencv
from google.colab.patches import cv2_imshow # Import the cv2_imshow function from google.colab.patches
cv2_imshow(img) # Display the image using cv2_imshow
```

```python
#save an image to any storage device by OpenCV.
cv2.imwrite(r"D:\albert-einstein_gray_opencv.jpg",img) #'cv2.imwrite' method is used to save an image to any storage device.
```
